# key factors that influence US home prices nationally

#### There can be various key factors which can affect the house price in U.S. some of the factor which i found important during my research i have explain them below 

1] Economic Growth: The demand for homes is linked to economic growth. When the economy grows, people have more money to spend on housing, driving up prices. In contrast, during a recession, fewer people can afford homes, leading to lower prices.

2] Unemployment: Rising unemployment makes it harder for people to afford homes. Even the fear of job loss can deter people from entering the real estate market.

3] Interest Rates: Higher interest rates increase mortgage costs, reducing demand for home purchases. This makes renting more attractive, especially for homeowners with adjustable mortgage rates.

4] Customer Trust: Confidence plays a role in taking the risk of getting a mortgage. If people fear a decrease in house prices, they may delay purchasing.

5] Supply and Demand: Shortages in housing supply drive prices up, while oversupply can lead to price decreases.

6] Home Sales Reflect Economy: Home sales are directly tied to economic stability and growth. Economic slowdown limits cash availability, making the housing market less accessible.

7] GDP Impact: As the Gross Domestic Product (GDP) increases, US house prices tend to rise, and vice versa.

8] Supply and Demand Laws: Like any industry, the housing market follows the laws of supply and demand. More buyers than sellers increase demand and make houses more expensive.

9] Advancement: Predicting future land benefits contributes to increasing property values over time.

10] Compliance: Consistent land use regulations maintain property values, while inconsistent regulations may negatively impact value.

11] Competition: Competing developments can affect supply and demand, influencing housing prices.

12] External Factors: Changes in the surrounding environment, such as new constructions or closures, can impact home values.

13] Surplus Productivity: The difference between cost and sale price affects the value of homes.

14] Location: Desirable areas tend to appreciate in value over time.

15] Neighborhood Factors: Proximity to highways, attractions, hospitals, and schools can influence house prices.

16] Property Taxes: Property taxes vary based on local factors, affecting overall housing costs.

17] Crime Rate: Areas with higher crime rates may have lower house prices due to lower desirability.

18] Government Laws: Government policies, such as tax deductions and tariffs on housing materials, can impact home prices.

19] Mortgage Availability: During economic booms, banks are more willing to lend mortgages, affecting the affordability of homes.

20] Area Desirability: Factors like good school districts and trendy neighborhoods can make housing prices rise, even when the national trend is different.

# Importing Librarie 

```python
#Importing libraries
import numpy as np
import pandas as pd
```

```python
#Importing different different csv files which contains our data 
label= pd.read_csv('y home price.csv')

#change data type of date column into datetime formate
label["DATE"] = pd.to_datetime(label["DATE"])

# Creat "Year" and "Month" columns
label["Year"] = pd.DatetimeIndex(label["DATE"]).year
label["Month"] = pd.DatetimeIndex(label["DATE"]).month
print(label.shape)
label.head()
```

```python
#Importing Unemployment Rate Data csv file
unemployment= pd.read_csv('unemployment rate.csv')
print(unemployment.shape)
unemployment.head()
```

```python
#Importing Per Capita GDP Data csv file 
gdp = pd.read_csv('per cap gdp.csv', names = ["DATE", "Per_Capita_GDP"], skiprows = 1)
print(gdp.shape)
gdp.head()
```

The data is quarterly. We will impute for other months using linear interpolation after while creating the final dataframe combining all the data.

```python
#Importing Interest Rate Data csv file 
int_rate = pd.read_csv('interest rate.csv')
print(int_rate.shape)
int_rate.head()
```

```python
#Importing construction material price Data csv file
cons_price = pd.read_csv('construction material price.csv', names = ["DATE", "Con_Materials"], skiprows = 1)
print(cons_price.shape)
cons_price.head()
```

```python
#Importing Consumer Price Index Data csv file
CPI = pd.read_csv('consumer price index.csv', names = ["DATE", "CPI"],skiprows = 1)
print(CPI.shape)
CPI.head()
```

```python
#Importing Monthly supply of new house Data csv file
new_house = pd.read_csv('monthly supply of new houses in us.csv', names = ["DATE", "Houses"], skiprows = 1)
print(new_house.shape)
new_house.head()
```

```python
#Importing Housing Subsidies csv Data file
house_subside = pd.read_csv('houseing subsides.csv', names = ["DATE", "Subsidy"], skiprows = 1)
print(house_subside.shape)
house_subside.head()
```

```python
#Importing Number of households csv Data file 
num_households = pd.read_csv('total household.csv', names = ["DATE", "Num_Households"], skiprows = 1)
print(num_households.shape)
num_households.head()
```

```python
#Importing  Working age population csv Data file
working_age = pd.read_csv('working age population.csv', names = ["DATE", "Working_Population"], skiprows = 1)
print(working_age.shape)
working_age.head()
```

```python
#Importing Population above 65 Data csv file
old_pop = pd.read_csv('population above 65.csv', names = ["DATE", "old_percent"], skiprows = 1)
print(old_pop.shape)
old_pop.head()
```

```python
#Importing Real Median Household Income csv Data file 
household_income = pd.read_csv('real median household income.csv', names = ["DATE", "Income"], skiprows = 1)
print(household_income.shape)
household_income.tail()
```

```python
# Merging Per Capita GDP (Quarterly data)
gdp["DATE"] = pd.to_datetime(gdp["DATE"])
label = pd.merge(label,gdp, how = "left")
label.head()
```

```python
print(label)
```

```python
# Concating dataframes having monthly data to create one dataframe
df = pd.DataFrame()
df_bymonth = [label, working_age, new_house, CPI, unemployment, cons_price, int_rate]
for df in df_bymonth:
    df1["DATE"] = pd.to_datetime(df1["DATE"])
    df1 = df1.set_index("DATE")
    df = pd.concat([df,df1], axis = 1)
print(df.shape)
df.head()
```

```python
# Merging other dataframes 
other_col = [num_households, household_income, house_subside, old_pop]
for df2 in other_col:
    if "Year" not in df2.columns:
        df2["Year"] = pd.DatetimeIndex(df2["DATE"]).year
        df2.set_index("DATE", inplace = True)
        df = pd.merge(df, df2, how = "left", on = "Year")
    else:
        df2.set_index("DATE", inplace = True)
        df = pd.merge(df, df2, how = "left", on = "Year")
df["DATE"] = label["DATE"]
df.set_index("DATE", inplace = True)
df.head()
```

```python
df.isna().sum()
```

Column Per_Capita_GDP has higher missing value so we will use linear interpolation to deal with missing value here & for other column we can simply drop the missing data

```python
# Filling missing values in the Per_Capita_GDP column using interpolation
df["Per_Capita_GDP"] = df["Per_Capita_GDP"].interpolate()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
```
